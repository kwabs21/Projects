```python
# This project is still being worked on

import streamlit as st
import pandas as pd

st.header('Lithology Prediction App')
# st.write('Lithology Prediction App')

# Creating the input section
st.sidebar.header('User Input Features')

file_type = st.sidebar.selectbox('Choose file type', ('csv', 'excel'))

# Input 1 (Dropping csv)
if file_type == 'csv':
    uploaded_file = st.sidebar.file_uploader('Upload your CSV file', type=['csv'])
    df = pd.read_csv(uploaded_file)
else:
    uploaded_file = st.sidebar.file_uploader('Upload your EXCEL file', type=['xlsx'])
    df = pd.read_excel(uploaded_file)
# uploaded_file = st.sidebar.file_uploader('Upload your CSV file', type=['csv','xlsx'])

st.subheader('Data')
st.dataframe(df)

# st.write(df.columns)

# Select features and labels
features = st.multiselect('Features', list(df.columns))

labels = st.multiselect('Labels', list(df.columns))
```
